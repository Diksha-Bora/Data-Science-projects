**DATA MODELLING**

*DATA SACLING USING MIN MAX*

```python
min = data["Annual Income (k$)"].min()
max = data["Annual Income (k$)"].max()
data["Annual Income (k$) Scaled"] = (data["Annual Income (k$)"]-min)*100/ (max - min)
data.head()
```

```python
min = data["Spending Score (1-100)"].min()
max = data["Spending Score (1-100)"].max()
data["Spending Score (1-100) Scaled"] = (data["Spending Score (1-100)"]-min)*100/ (max - min)
data.head()
```

```python
data["Annual Income (k$) Scaled"] = data["Annual Income (k$) Scaled"].round()
data["Spending Score (1-100) Scaled"] = data["Spending Score (1-100) Scaled"].round()
```

```python
data = data.drop(["Annual Income (k$)","Spending Score (1-100)","CustomerID"], axis =1)
data.head()
```

```python
plt.figure(figsize=[15,10])
df = data.groupby("Age").mean()
df.plot.area(stacked = False)
plt.ylabel('Score')
plt.title('Spending score vs Income score')
plt.show()
```

```python
data["Gender"] = data["Gender"].map({"Male":0, "Female":1})
```

```python
data
```

```python
from sklearn.cluster import KMeans
inertia = []
num_of_clusters = []
for i in range(1,20):
  km = KMeans(n_clusters= i)
  km.fit(data)
  inertia.append(km.inertia_)
  num_of_clusters.append(i)
```

```python
plt.figure(figsize=[10,8])
plt.plot(num_of_clusters , inertia, marker =".", markersize=20)
```

**KMeans**

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters = 6)
labels = km.fit_predict(data)
```

```python
labels
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(data, labels)
```

**KMeans++**

```python
from sklearn.cluster import KMeans
kmp = KMeans(n_clusters = 6, init = "k-means++",n_init = 15, max_iter = 350)
kmp.fit(data)
```

```python
labels1 = kmp.predict(data)
```

```python
labels1
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(data, labels1)
```

**Agglomerative Clustering**

```python
from sklearn.cluster import AgglomerativeClustering
hier = AgglomerativeClustering(n_clusters = 6)
labels2 = hier.fit_predict(data)
```

```python
labels2
```

***DENDROGRAM***

```python
from scipy.cluster.hierarchy import dendrogram, linkage
plt.figure(figsize=[20,15])
merg = linkage(data, method = "ward")
dendrogram(merg, leaf_rotation=90)
plt.title("Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Euclidean Distance")
plt.show()
```

**DBSCAN = Density-based spatial clustering of applications with noise**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(data)
x = scaler.transform(data)
```

```python
from sklearn.cluster import DBSCAN
ds  = DBSCAN(eps=0.8, min_samples = 10)
labels3 = ds.fit_predict(x)
```

```python
labels3
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(x, labels3)
```
